Here, i have done ` Data Analsys ` part of popular Github Repositary.

```python
import pandas as pd
data = pd.read_csv("Github_data.csv")
data.sample(5)
```

```python
del_list = ['Unnamed: 0','Unnamed: 0.1']
data.drop(del_list,axis=1,inplace=True)
data.sample(5)
```

```python
data.head()
```

```python
data.info()
```

```python
oolumns_name = data.columns
oolumns_name
```

```python
# Rename the columns
new_column = ['Topic', 'Name', 'User', 'Star', 'Fork', 'Watch', 'Issue',
       'Pull_requests', 'Projects', 'Topic_tag', 'Discription_text',
       'Discription_url', 'Commits', 'Branches', 'Packages', 'Releases',
       'Contributers', 'License', 'Url']

data = data.rename(columns= dict(zip(oolumns_name,new_column)))
data.head(5)
```

```python
data['Topic'].value_counts().plot(kind='bar')
```

```python
data['License'].value_counts()
```

```python
data[data['License'] == 'BSD-3-Clause']
```

```python
data.info()
```

```python
data['Star'] = data['Star'].apply(lambda x : float(x.rstrip('k'))*1000 if x.endswith('k') else float(x))
data['Fork'] = data['Fork'].apply(lambda x : float(x.rstrip('k'))*1000 if x.endswith('k') else float(x))
data['Watch'] = data['Watch'].apply(lambda x : float(x.rstrip('k'))*1000 if x.endswith('k') else float(x))
data.head(5)
```

```python
data['Issue'] = data['Issue'].apply(lambda x : x.replace(',','') if ',' in x else x)
data['Commits'] = data['Commits'].apply(lambda x : x.replace(',','') if ',' in x else x)
data.head()
```

```python
data.info()
```

```python
cols = ['Issue','Pull_requests','Commits','Contributers']
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce', axis=1)
data.dtypes
```

```python
data.describe()
```

```python
new_data = data.groupby('Topic').mean().reset_index()
new_data
```

### Question:- Try to understand contribution graph

```python
corr_df = data.dropna(axis =0, subset = ['Issue','Pull_requests','Contributers','Commits'])[['Issue','Pull_requests','Contributers','Commits']]
```

```python
corr_df.corr()
```

```python
# Correlation between the contribution column
import seaborn as sns
sns.heatmap(corr_df.corr() , linewidths =0.1, vmax =1.0, square = True, linecolor ='white' , annot = True, cmap ='summer')
```

#### Compare above result with result of 100 most popular repo

```python
corr_100df = data.nlargest(n=100,columns=['Star'])[['Issue','Pull_requests','Contributers','Commits']]
```

```python
corr_100df.corr()
```

### Analysis of Tags

```python
data['Topic_tag'].sample(10)
```

```python
import ast
from collections import Counter

topic_tags = data['Topic_tag'].apply(lambda x: ast.literal_eval(x)).tolist()

all_tags = [tag for topic in topic_tags for tag in topic]

tags_dict = Counter(all_tags)
```

```python
toptags_df = pd.DataFrame(tags_dict.most_common(15), columns=['Name of the Tag','Count'])
```

```python
sns.barplot(data=toptags_df, x='Name of the Tag', y='Count', palette='autumn')
plt.xticks(rotation=90)
```

The most famous Tags are associated with ` Data Science` .
